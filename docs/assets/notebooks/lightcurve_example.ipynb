# Lightcurve example
A simple notebook (only two functions!) to show how to construct a point-source point-lens (PSPL) lightcurve.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def A(u):
    """
    The magnification equation as a function of the
    dimensionless source-lens projected separation u.
    """
    A = (u**2 + 2)/(u * np.sqrt(u**2 + 4))
    return A
```

```python
def u(t, tE, u0, t0): 
    """
    The dimensionless source-lens projected separation u
    as a function of time.

    t : time.
    tE : Einstein crossing time.
    u0 : impact parameter.
    t0 : time of closest approach.
    """
    u = np.sqrt(u0**2 + ((t - t0)/tE)**2)
    return u
```

```python
fig = plt.figure(1, figsize=(6,6))
t = np.linspace(-50, 50, 1000)
plt.plot(t, A(u(t, 10, 0.1, 0)), label = '$u_0 = 0.1$')
plt.plot(t, A(u(t, 10, 0.5, 0)), label = '$u_0 = 0.5$')
plt.plot(t, A(u(t, 10, 1, 0)), label = '$u_0 = 1$')
plt.plot(t, A(u(t, 10, 3, 0)), label = '$u_0 = 3$')
plt.xlabel('Time (days)')
plt.ylabel('Magnification')
plt.title('Effect of changing $u_0$')
plt.legend()
plt.show()
```

```python
fig = plt.figure(1, figsize=(6,6))
t = np.linspace(-50, 50, 1000)
plt.plot(t, A(u(t, 10, 1, -10)), label = '$t_0 = -10$ days')
plt.plot(t, A(u(t, 10, 1, 0)), label = '$t_0 = 0$ days')
plt.plot(t, A(u(t, 10, 1, 20)), label = '$t_0 = 20$ days')
plt.xlabel('Time (days)')
plt.ylabel('Magnification')
plt.title('Effect of changing $t_0$')
plt.legend()
plt.show()
```

```python
fig = plt.figure(1, figsize=(6,6))
t = np.linspace(-50, 50, 1000)
plt.plot(t, A(u(t, 1, 1, 0)), label = '$t_E = 1$ days')
plt.plot(t, A(u(t, 3, 1, 0)), label = '$t_E = 3$ days')
plt.plot(t, A(u(t, 10, 1, 0)), label = '$t_E = 10$ days')
plt.xlabel('Time (days)')
plt.ylabel('Magnification')
plt.title('Effect of changing $t_E$')
plt.legend()
plt.show()
```
